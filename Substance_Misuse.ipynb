# Create a dataset with "Crime Name2" containing "Drug/Narcotic Violations" without empty values and missing "Crime Name3"

```python
df = df[df["Crime Name2"].str.contains("Drug/Narcotic Violations", na=False) & df["Crime Name3"].notna()]
```

```python
df
```

# Use .str.extract() with regex to capture the text between two dashes

```python
df["Substance"] = df["Crime Name3"].str.extract(r"-\s*([^-]+?)\s*-")
```

```python
df
```

```python
df["Misuse"] = df["Crime Name3"].str.extract(r"-\s*([^-]+)$")
```

```python
df
```

```python
df["Misuse"].value_counts()
```

# Create a dataset with "Misuse" containing unidentified misues all converted to "Other"

## Replace rows in 'Column' where 'DESCRIBE' is present

```python
df["Misuse"] = df["Misuse"].apply(lambda x: 'OTHER' if 'DESCRIBE' in x else x)
df["Misuse"].value_counts()
```

```python
df.to_csv("Crime_Dataset_Substances.csv", index=False)
```
